```python
hdu = fits.open("hst_results_nd.fits")
```

```python
hdr = hdu[0].header
data = hdu[0].data
```

```python
t = Table.read("hst_results_nd.fits")
ra = t["RA"]
dec = t["DEC"]
rv = t['Rv_p50']
# t[0].colnames
```

```python
plt.figure()
# plt.plot(ra, dec,'.b', ls='')
cb = plt.scatter(ra, dec, c = rv,marker='s', cmap = 'berlin', vmin=0, vmax=1.5) 
plt.colorbar(cb, label='R(V)')
plt.xlabel("RA", fontsize=14)
plt.ylabel('DEC', fontsize=14)
plt.savefig("lmc_rv_spacial.png")
```

```python
dist = t["distance_p50"]
```

```python
plt.figure()
plt.hist(dist/1000, bins = 25, color = 'k')
plt.xlabel("Distance[kpc]", fontsize = 14)
plt.ylabel("N")
d_mean = np.mean(dist/1000)
print(d_mean)
plt.axvline(d_mean, color = 'r', ls = '--', lw = 1)
plt.text(58, 270, '%s' % np.around(d_mean, decimals=2))
plt.savefig('lmc_hist.png')
plt.savefig('lmc_hist.pdf')
```
